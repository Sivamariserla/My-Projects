## **8.Checking the  heart rate pulse which contains in CHD ??**

```python
def heartrate_data(row):
  if row['heartRate']<=59:
    value='Low'
  elif row['heartRate']<80:
    value='Normal'
  else:
    value='High'
  return value
```

```python
df['heartRatelabel']=df.apply(heartrate_data,axis=1)
```

```python
df['heartRatelabel'].value_counts()
```

```python
x=pd.crosstab(df['heartRatelabel'],df['TenYearCHD'])
x.plot(kind='bar',width=0.8)
font1={'family':'serif','size':18,'color':'k'}
font2={'family':'serif','size':16,'color':'g'}
plt.xlabel('Heartrate pulse',fontdict=font2)
plt.ylabel('Counts',fontdict=font2)
plt.title('Heart rate pulses at risk in CHD',fontdict=font1)
plt.xticks(rotation=0)
plt.legend(['no_risk','at_risk'])
plt.show()
```
